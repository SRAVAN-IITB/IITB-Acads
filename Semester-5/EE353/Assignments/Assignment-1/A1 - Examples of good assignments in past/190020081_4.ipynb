5. Print a histogram of number missing values for each column sorted from low to high

```python
print(data.isnull().sum()) # finding number of null values in data
# creating a list of columns
col= ['Rowid','Country','State lgd code','State','District lgd code','District','Year','Type of residence namely rural or urban','Number of census houses','Houses having grass or thatch or bamboo etc. as material of wall','Houses having plastic or polythene as material of wall','Houses having mud or unburnt brick as material of wall','Houses having wood as material of wall','Houses having stone not packed with mortar as material of wall','Houses having stone packed with mortar as material of wall','Houses having galvanized iron (gi) sheets or metal or asbestos sheets as material of wall','Houses having burnt brick as material of wall','Houses having concrete as material of wall','Houses having any other material of wall']
null=[] # creating a list to store columns who have null values with frquency
for i in col: # for loop to look into each column
  j=0 # intialising index
  a=data[i].isnull().sum()# no of null values in above chosed column
  for j in range(a): # for value to add column name in list null
      null.append(i) # appending column to null
n_bins=11 # defining no. of bins for plot
fig, axs = plt.subplots(1, 1,figsize =(20, 10),tight_layout = True) # providing details for histogram
plt.hist(null, bins = n_bins) # priting histogram
plt.show() # command to show plot
```

All columns regarding number of houses have 5 missing values

6. Plot a histogram of number of missing values for each row sorted from low to high

```python
a=data.isnull().sum(axis=1) #storing null values row wise
print(a) # priting no. of null values
null=[] # creating a list to store rows having null values with frequency
n_bins=0 # intialise no. of bins for plot
i=0 # initialising variable for for loop
for i in range(rows): # for loop to traverse through all rows
  j=0 # initialising varibale for for loop
  b=a[i] # choosing null values in row i
  for j in range(b): # for loop to add null values in row i
      null.append(i) # adding rows to null 
      n_bins=n_bins+1 # increasing no. of bins according to values
fig, axs = plt.subplots(1, 1,figsize =(20, 10),tight_layout = True) #specialising figure details
plt.hist(null, bins = n_bins) # plotting histogram
plt.show() # showing plotted figure
```

5 of the rows have 11 missing values corresponding to columns for number of houses

7. If any column’s datatype or values needs to be cleaned, then do so

10. For all pairs of continuous variables, plot the scatter plot and show color-coded correlation matrix

```python
# creating a list to store columns which are numerical type
col_cont= ['Rowid','State lgd code','District lgd code','Year','Number of census houses','Houses having grass or thatch or bamboo etc. as material of wall','Houses having plastic or polythene as material of wall','Houses having mud or unburnt brick as material of wall','Houses having wood as material of wall','Houses having stone not packed with mortar as material of wall','Houses having stone packed with mortar as material of wall','Houses having galvanized iron (gi) sheets or metal or asbestos sheets as material of wall','Houses having burnt brick as material of wall','Houses having concrete as material of wall','Houses having any other material of wall']
plt.figure(figsize=(10,8), dpi= 80) # specifying details of figures to be plotted
sb.pairplot(data,vars=col_cont,kind="scatter", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5)) # creating pairplot of scatter type for all numerical type column 
plt.show() # showing the pairplot
```

some of the variables have nice linear relation where some are patternless

```python
# Plot
df = pd.DataFrame() # creating a new dataframe
for i in col_cont:# using for loop to select colimns from col_cont list
  df[i]=data[i] # adding selected cloumns to df dataframe
plt.figure(figsize=(12,12), dpi= 80) # specifying figure details
sb.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True) # creating heatmap having correlation values for column pairs

# Decorations
plt.title('Correlogram of NDAP Report', fontsize=22) # specifying title for plot
plt.xticks(fontsize=12) # specifying font size for ticks at x-axis
plt.yticks(fontsize=12) # specifying font size for ticks at y-axis
plt.show() # showing the figure
```

some correlation between number of houses and material brick and concrete can be seen. which is also true in real world

11. Check if the sum of columns J through S matches with column I for all rows

```python
# created a list for columns that need to summed up according to the task
col_list= ['Houses having grass or thatch or bamboo etc. as material of wall','Houses having plastic or polythene as material of wall','Houses having mud or unburnt brick as material of wall','Houses having wood as material of wall','Houses having stone not packed with mortar as material of wall','Houses having stone packed with mortar as material of wall','Houses having galvanized iron (gi) sheets or metal or asbestos sheets as material of wall','Houses having burnt brick as material of wall','Houses having concrete as material of wall','Houses having any other material of wall']
# sum specific columns
data['Sum'] = data[col_list].sum(axis=1) # creating a new column Sum for storing values of sum for col_list columns
```

```python
data.head() # priting head of column to look at newly created variable
```

```python
if data['Number of census houses'].all()==data['Sum'].all(): # using if statement to find out if Sum is equal to number of census houses
    print('Yes') # print yes if it is
else: # else for alternative action
    print('NO') # print no if it isn't
```

sum of all clomns is equal to total number which is what expected from a good balanced data

12. Divide columns J through S by column I and store the in new columns as percent (multiply by 100)

```python
for i in col_list: # for loop to select colums form col_list
  data[i+' new']=data[i]*100/data['Number of census houses'] # creating new colums having percentage values
```

15. For the new columns, plot pair-wise scatter plots and correlation matrix

```python
col_list_new = [x + ' new' for x in col_list] # creating a new list for all newly created columns from columns in col_list
plt.figure(figsize=(10,8), dpi= 80) #specifying figure details
sb.pairplot(data,vars=col_list_new, kind="scatter", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5)) # creating pairplot
plt.show() # showing pairplot
```

all of the data have nearly similar data. we can give a triangular type area which contains all of the data

```python
# Plot
df_new = pd.DataFrame() # creating a new dataframe
for i in col_list_new: # selecting columns from col_list_new
  df_new[i]=data[i] # adding new columns to dataframe df_new
plt.figure(figsize=(12,12), dpi= 80) # specifying figure details
# creating heatmap containg correlation values for pair of columns
sb.heatmap(df_new.corr(), xticklabels=df_new.corr().columns, yticklabels=df_new.corr().columns, cmap='RdYlGn', center=0, annot=True)

# Decorations
plt.title('Correlogram of NDAP Report for new columns', fontsize=22) # creating title for variable
plt.xticks(fontsize=12) # specify fontsize for xticks
plt.yticks(fontsize=12) # specify fontsize for yticks
plt.show() # showing heatmap
```

bamboo and brick have negative correlation which indicates these two aren't used together which is somehow right in real world also. where bamboo and plastic have positive correlation which indicates use of these two together.
