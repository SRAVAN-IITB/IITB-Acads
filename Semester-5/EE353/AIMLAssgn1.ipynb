# Extracting ZIP Codes from 'SITE ADDRESS'

Since the Affordable Housing Projects dataset does not have a dedicated ZIP code column, we need to extract ZIP codes from the `'SITE ADDRESS'` column. Here's how we achieve that:

1. **Define Extraction Function**: We create a function named `extract_zip_code` that uses a regular expression to identify and extract ZIP codes from a given address. The regular expression `\b\d{5}(?:-\d{4})?\b` is designed to match standard 5-digit ZIP codes and ZIP+4 codes.

2. **Apply Extraction Function**: We apply the `extract_zip_code` function to each value in the `'SITE ADDRESS'` column. This creates a new column called `'ZIP_CODE'` in the DataFrame, which contains the extracted ZIP codes.

3. **Convert to String**: To ensure consistency and ease of use, we convert the new `'ZIP_CODE'` column to a string type.

4. **Verify Extraction**: Finally, we display the first few rows of the DataFrame to check if the ZIP codes have been correctly extracted from the addresses.

This process helps us create a usable ZIP code column from the available address information, enabling further analysis.



````python
import re

# Function to extract ZIP code from address
def extract_zip_code(address):
    match = re.search(r'\b\d{5}(?:-\d{4})?\b', address)
    return match.group(0) if match else None

# Apply the function to the 'SITE ADDRESS' column
housing_projects_df['ZIP_CODE'] = housing_projects_df['SITE ADDRESS'].apply(extract_zip_code)

# Ensure 'ZIP_CODE' is treated as a string (object)
housing_projects_df['ZIP_CODE'] = housing_projects_df['ZIP_CODE'].astype(str)

# Display the first few rows to verify
print(housing_projects_df[['SITE ADDRESS', 'ZIP_CODE']].head())
````

````python
# Plot histograms for continuous variables
def plot_continuous_histograms(df, columns):
    for col in columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[col], bins=30, kde=True)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Density')
        plt.show
````

````python
# Plot side-by-side boxplot for each variable
def plot_side_by_side_boxplots(df, columns):
    df_melted = df[columns].melt()
    plt.figure(figsize=(12, 8))
    sns.boxplot(x='variable', y='value', data=df_melted)
    plt.xticks(rotation=90)
    plt.title('Side-by-Side Boxplots for Selected Variables')
    plt.show()
````

````python
# Plot the correlation matrix
def plot_correlation_matrix(df):
    corr = df.corr().abs()  # Absolute value of the correlation matrix
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=0, vmax=1)
    plt.title('Correlation Matrix')
    plt.show()
````

# Hypothesis Formulation

**Research Question:**
Does the presence of affordable housing projects in a ZIP code impact the health inspection scores of restaurants in that ZIP code?

**Null Hypothesis (H0):**
There is no relationship between the statistics of affordable housing projects and the health inspection scores of restaurants in the same ZIP code.

**Alternative Hypothesis (H1):**
There is a significant relationship between the statistics of affordable housing projects and the health inspection scores of restaurants in the same ZIP code.

# Reasonable Alternative Hypothesis

**Alternative Hypothesis (H1):**
- The presence and characteristics of affordable housing projects in a ZIP code (e.g., number of units, funding, job creation) are significantly related to the health inspection scores of restaurants within that ZIP code.

In other words:
- Higher or lower values of predictor variables (such as `SITE UNITS`, `PROJECT TOTAL UNITS`, `LAHD FUNDED`, etc.) in a ZIP code are associated with better or worse health inspection scores (`AVG_SCORE`, `MIN_SCORE`, `MAX_SCORE`) for restaurants in that same ZIP code.

This hypothesis allows you to investigate whether any observed variations in affordable housing statistics correlate with variations in restaurant health scores, suggesting a potential impact or association.


````python
import pandas as pd
import statsmodels.api as sm
from scipy import stats

# Load the joined dataset
joined_df = pd.read_csv('/content/drive/MyDrive/Semester 5 Jargon/EE353 Assignment 1/Housing_Restaurant_Joined_by_ZIP.csv')

# Drop rows with missing values for the analysis
joined_df = joined_df.dropna(subset=['AVG_SCORE', 'SITE UNITS'])

# Define predictor and predicted variables
predictor_var = 'SITE UNITS'
predicted_var = 'AVG_SCORE'

# Prepare the data for regression analysis
X = joined_df[[predictor_var]]
y = joined_df[predicted_var]

# Add constant to the predictor variable to include an intercept in the model
X = sm.add_constant(X)

# Perform linear regression
model = sm.OLS(y, X).fit()

# Print the summary of the regression analysis
print(model.summary())

# Perform hypothesis test
# H0: There is no relationship between the number of site units and average score.
# H1: There is a significant relationship between the number of site units and average score.

# Extract p-value from the model summary
p_value = model.pvalues[predictor_var]

# Define significance level
alpha = 0.05

# Test the hypothesis
if p_value < alpha:
    print(f"The p-value is {p_value:.5f}. Since it is less than the significance level of {alpha}, we reject the null hypothesis.")
else:
    print(f"The p-value is {p_value:.5f}. Since it is greater than the significance level of {alpha}, we fail to reject the null hypothesis.")
````

````python
import pandas as pd
import statsmodels.api as sm

# Load the joined dataset
joined_df = pd.read_csv('/content/drive/MyDrive/Semester 5 Jargon/EE353 Assignment 1/Housing_Restaurant_Joined_by_ZIP.csv')

# Define predictor and response variables
predictor_vars = ['SITE UNITS', 'PROJECT TOTAL UNITS', 'JOBS', 'LAHD FUNDED', 'TDC']
response_vars = ['AVG_SCORE', 'MIN_SCORE', 'MAX_SCORE']

# Significance level
alpha = 0.05

# Perform Linear Regression for each combination of predictor and response variable
for predictor in predictor_vars:
    for response in response_vars:
        # Drop rows with missing values in the predictor or response variable
        df_clean = joined_df[[predictor, response]].dropna()

        # Define the predictor (X) and response (y)
        X = df_clean[[predictor]]
        y = df_clean[response]

        # Add a constant (intercept) to the predictor variable
        X = sm.add_constant(X)

        # Perform Linear Regression
        model = sm.OLS(y, X).fit()

        # Extract p-value
        p_value = model.pvalues[predictor]

        # Print the results
        print(f'\nLinear Regression Summary for {predictor} vs {response}:')
        print(model.summary())

        # Classify based on p-value
        if p_value < alpha:
            print(f"The relationship between {predictor} and {response} is statistically significant (p-value = {p_value:.4f}).")
        else:
            print(f"The relationship between {predictor} and {response} is not statistically significant (p-value = {p_value:.4f}).")
````

````python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load the datasets from the provided file paths
rainfall_data = pd.read_csv('/content/drive/MyDrive/Semester 5 Jargon/EE353 Assignment 1/RF_AI_1901-2021.csv')
temperature_data = pd.read_csv('/content/drive/MyDrive/Semester 5 Jargon/EE353 Assignment 1/TEMP_ANNUAL_SEASONAL_MEAN.csv')

# Check column names for both datasets
print("Rainfall Data Columns:")
print(rainfall_data.columns)

print("\nTemperature Data Columns:")
print(temperature_data.columns)

# Display the first few rows of each dataset
print("Rainfall Data:")
print(rainfall_data.head())

print("\nTemperature Data:")
print(temperature_data.head())

# Calculate annual rainfall by summing the monsoon months (JUN, JUL, AUG, SEP)
rainfall_data['Annual Rainfall'] = rainfall_data[['JUN', 'JUL', 'AUG', 'SEP']].sum(axis=1)

# Merge the datasets on the 'YEAR' column
merged_data = pd.merge(rainfall_data[['YEAR', 'Annual Rainfall']], temperature_data[['YEAR', 'ANNUAL']], on='YEAR')

# Convert the columns to numeric, forcing errors to NaN
merged_data['ANNUAL'] = pd.to_numeric(merged_data['ANNUAL'], errors='coerce')
merged_data['Annual Rainfall'] = pd.to_numeric(merged_data['Annual Rainfall'], errors='coerce')

# Drop rows with missing values
merged_data.dropna(inplace=True)

# Perform exploratory data analysis (EDA)
print("Dataset Info:")
print(merged_data.info())

print("\nDataset Description:")
print(merged_data.describe())

# Plot the data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(merged_data['Annual Rainfall'], kde=True, color='blue')
plt.title('Distribution of Annual Rainfall')

plt.subplot(1, 2, 2)
sns.histplot(merged_data['ANNUAL'], kde=True, color='red')
plt.title('Distribution of Mean Annual Temperature')

plt.tight_layout()
plt.show()

# Scatter plot of Annual Rainfall vs Mean Temperature
plt.figure(figsize=(8, 6))
sns.scatterplot(x='ANNUAL', y='Annual Rainfall', data=merged_data)
plt.title('Annual Mean Temperature vs Annual Rainfall')
plt.xlabel('Annual Mean Temperature (Â°C)')
plt.ylabel('Annual Rainfall (mm)')
plt.show()

# Perform a Pearson correlation test
corr, p_value = pearsonr(merged_data['ANNUAL'], merged_data['Annual Rainfall'])

print(f"Pearson Correlation: {corr}")
print(f"P-value: {p_value}")

# Interpretation of results
if p_value < 0.05:
    print("There is a statistically significant correlation between the mean temperature and annual rainfall.")
else:
    print("There is no statistically significant correlation between the mean temperature and annual rainfall.")
````
